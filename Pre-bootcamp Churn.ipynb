```python
train['TENURE'].replace(train['TENURE'].unique(),range(1,(train['TENURE'].nunique()+1)),inplace=True)
train['TENURE']
```

```python
train['MRG'].replace(train['MRG'].unique(),range(1,(train['MRG'].nunique()+1)),inplace=True)
train['MRG']
```

```python
sum(train['MRG']!=1)
```

```python
train.drop(['REGION','user_id','MRG'],axis=1,inplace=True)
train.fillna(0,inplace=True)
```

```python
train['TOP_PACK']=train['TOP_PACK'].apply(lambda x:str(x).split('=')[0])
#train['TOP_PACK']=train['TOP_PACK'].apply(lambda x:str(x).split()[0])
train['TOP_PACK'].unique()
```

```python
train['TOP_PACK'].replace(train['TOP_PACK'].unique(),range(1,(train['TOP_PACK'].nunique()+1)),inplace=True)
```

```python
train['TOP_PACK']=train['TOP_PACK'].astype(str)
train['TOP_PACK'].replace('0','None',inplace=True)
train['TOP_PACK']
```

```python
#sum(train[train['TOP_PACK']=='None']['FREQ_TOP_PACK']!=0)
```

```python
train['FREQ_TOP_PACK']=pd.qcut(train['FREQ_TOP_PACK'],q=5,duplicates='drop',labels=[1,2,3])
```

```python
train['FREQUENCE_RECH']=pd.qcut(train['FREQUENCE_RECH'],q=5,duplicates='drop',labels=[1,2,3,4])
```

```python
train['FREQUENCE']=pd.qcut(train['FREQUENCE'],q=5,duplicates='drop',labels=[1,2,3,4])
```

```python
train['REGULARITY']=pd.qcut(train['REGULARITY'],q=4,duplicates='drop',labels=[1,2,3,4])
```
